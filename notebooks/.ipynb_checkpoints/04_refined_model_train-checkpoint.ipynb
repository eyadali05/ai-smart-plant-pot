```python
# Cell 1 — imports & paths
import os, random, math, json, shutil
from pathlib import Path
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from PIL import Image, ImageFilter, ImageEnhance
import cv2
from tensorflow.keras import mixed_precision

mixed_precision.set_global_policy('mixed_float16')

# repo structure (adjust if different)
ROOT = Path.cwd().parent  # if notebook is in notebooks/, parent is repo root
processed_root = ROOT / "data_processed"
models_dir = ROOT / "models"
models_dir.mkdir(exist_ok=True)
augment_samples_dir = models_dir / "augment_samples"
augment_samples_dir.mkdir(exist_ok=True)

IMG_SIZE = 192
BATCH_SIZE = 32
AUTOTUNE = tf.data.AUTOTUNE

# load class names from folder names
classes = sorted([p.name for p in processed_root.iterdir() if p.is_dir()])
num_classes = len(classes)
print("Classes:", num_classes)
```

```python
# Cell 2 — helpers
def clean_path(p):
    return str(p).replace("\\","/")

def list_files_for_split(split="train"):
    files, labels = [], []
    for idx, cls in enumerate(classes):
        p = processed_root / cls / split
        if not p.exists():
            continue
        for f in p.glob("*"):
            files.append(clean_path(f))
            labels.append(idx)
    return files, labels

def decode_img(path):
    # read file, decode and return float32 [0,1]
    img = tf.io.read_file(path)
    img = tf.image.decode_jpeg(img, channels=3)
    img = tf.image.convert_image_dtype(img, tf.float32)  # [0,1]
    return img

def resize_and_center_crop(img, size=IMG_SIZE):
    img = tf.image.resize(img, [size, size])
    return img
```

```python
import os
import io
from PIL import Image, UnidentifiedImageError
from pathlib import Path

print("Scanning and cleaning dataset...")

bad_files = []
converted_files = 0
checked_files = 0

def is_image_ok(path):
    """Try to open image — returns False if corrupted."""
    try:
        img = Image.open(path)
        img.verify()  # quick integrity check
        return True
    except Exception:
        return False


def convert_to_jpeg(path):
    """Convert unsupported formats to JPEG."""
    global converted_files
    try:
        img = Image.open(path).convert("RGB")
        new_path = path.with_suffix(".jpg")
        img.save(new_path, "JPEG", quality=95)
        os.remove(path)
        converted_files += 1
        return new_path
    except Exception:
        return None


root = processed_root  # from previous cell

# Walk through all files
for cls in classes:
    for split in ["train", "val", "test"]:
        folder = root / cls / split
        if not folder.exists():
            continue
        
        for f in folder.iterdir():
            checked_files += 1
            if not f.is_file():
                continue

            # Skip known non-image files
            if f.suffix.lower() not in [".jpg", ".jpeg", ".png", ".bmp"]:
                # Try converting to jpeg
                new = convert_to_jpeg(f)
                if new is None:
                    bad_files.append((str(f), "Unsupported format"))
                continue

            # Check image integrity
            if not is_image_ok(f):
                # try reloading and converting to clean JPEG
                try:
                    img = Image.open(f).convert("RGB")
                    img.save(f, "JPEG", quality=90)
                except Exception:
                    bad_files.append((str(f), "Corrupted file"))
                    continue

print("=== CLEANING SUMMARY ===")
print("Checked files:", checked_files)
print("Converted files:", converted_files)
print("Corrupted files found:", len(bad_files))

# Remove corrupted files
for f, reason in bad_files:
    os.remove(f)
    print("Deleted:", f, "| Reason:", reason)

print("✅ Dataset cleaning completed successfully.")
```

```python
import numpy as np
import cv2
from PIL import Image, ImageEnhance, ImageFilter
from io import BytesIO

def augment_numpy(image_np, rng=None):
    if rng is None:
        rng = np.random.default_rng()

    img = Image.fromarray(image_np)

    # --- 1. Random crop/zoom (close-up plant variability) ---
    if rng.random() < 0.8:
        w, h = img.size
        scale = rng.uniform(0.85, 1.0)
        new_w, new_h = int(w*scale), int(h*scale)
        left = rng.integers(0, max(1, w-new_w))
        top  = rng.integers(0, max(1, h-new_h))
        img = img.crop((left, top, left+new_w, top+new_h))
        img = img.resize((IMG_SIZE, IMG_SIZE), Image.BILINEAR)
    else:
        img = img.resize((IMG_SIZE, IMG_SIZE), Image.BILINEAR)

    # --- 2. Brightness & contrast ---
    if rng.random() < 0.6:
        img = ImageEnhance.Brightness(img).enhance(rng.uniform(0.7, 1.25))
    if rng.random() < 0.6:
        img = ImageEnhance.Contrast(img).enhance(rng.uniform(0.7, 1.25))

    # --- 3. Slight green tint (Pi Camera tendency) ---
    if rng.random() < 0.5:
        arr = np.array(img).astype(np.float32)
        arr[:,:,1] = np.clip(arr[:,:,1] * rng.uniform(0.98, 1.06), 0, 255)
        img = Image.fromarray(arr.astype(np.uint8))

    # --- 4. JPEG compression artifacts ---
    if rng.random() < 0.5:
        q = int(rng.uniform(35, 85))
        buffer = BytesIO()
        img.save(buffer, format="JPEG", quality=q)
        buffer.seek(0)
        img = Image.open(buffer).convert("RGB")

    # --- 5. Gaussian noise ---
    if rng.random() < 0.5:
        arr = np.array(img).astype(np.float32)
        noise = rng.normal(0, rng.uniform(5, 20), arr.shape)
        arr = np.clip(arr + noise, 0, 255)
        img = Image.fromarray(arr.astype(np.uint8))

    # --- 6. Motion blur ---
    if rng.random() < 0.3:
        k = rng.integers(3, 8)
        kernel = np.zeros((k, k))
        if rng.random() < 0.5:
            kernel[k//2, :] = 1.0 / k
        else:
            kernel[:, k//2] = 1.0 / k
        arr = np.array(img)
        arr = cv2.filter2D(arr, -1, kernel)
        img = Image.fromarray(arr)

    # --- 7. Vignette / shadows ---
    if rng.random() < 0.35:
        arr = np.array(img).astype(np.float32)
        h, w, _ = arr.shape
        Y, X = np.ogrid[:h, :w]
        dist = np.sqrt((X - w/2)**2 + (Y - h/2)**2)
        mask = 1 - 0.6 * (dist / np.sqrt(w*w + h*h))
        mask = np.clip(mask, 0.4, 1).astype(np.float32)
        arr *= mask[..., None]
        img = Image.fromarray(arr.astype(np.uint8))

    # --- 8. Median filter (light smoothing) ---
    if rng.random() < 0.3:
        img = img.filter(ImageFilter.MedianFilter(size=3))

    return np.array(img)
```

```python
def tf_augment(image, label):
    """
    Wraps augment_numpy using tf.py_function.
    image: float32 [0,1]
    returns: augmented image [0,1], same label
    """
    def _augment_py(img):
        img = img.numpy()                    # ✅ Tensor → NumPy
        img = (img * 255).astype(np.uint8)    # Float32 → uint8
        aug = augment_numpy(img)              # apply real augmentation
        aug = aug.astype(np.float32) / 255.0
        return aug

    aug_img = tf.py_function(
        _augment_py,
        [image],
        Tout=tf.float32
    )
    aug_img.set_shape([IMG_SIZE, IMG_SIZE, 3])
    return aug_img, label
```

```python
import tensorflow as tf

def safe_load_and_label(path, label):
    """Safely load an image file and return (image, label)."""
    def _load_py(path_str):
        try:
            img = tf.io.read_file(path_str)
            img = tf.image.decode_jpeg(img, channels=3)
            img = tf.image.convert_image_dtype(img, tf.float32)
            img = tf.image.resize(img, [IMG_SIZE, IMG_SIZE])
            return img, label
        except Exception as e:
            # return dummy tensor and invalid label
            return tf.zeros([IMG_SIZE, IMG_SIZE, 3], tf.float32), -1

    img, lbl = tf.py_function(
        func=_load_py,
        inp=[path],
        Tout=(tf.float32, tf.int32)
    )
    img.set_shape([IMG_SIZE, IMG_SIZE, 3])
    lbl.set_shape([])

    return img, lbl
```

```python
# ========= CLEAN DATASET PIPELINE (FINAL) =============

def get_paths_and_labels(class_names, split):
    paths = []
    labels = []
    for idx, cls in enumerate(class_names):
        p = processed_root / cls / split
        for f in p.iterdir():
            if f.is_file():
                paths.append(str(f))
                labels.append(idx)
    return paths, labels

train_paths, train_labels = get_paths_and_labels(classes, "train")
val_paths,   val_labels   = get_paths_and_labels(classes, "val")
test_paths,  test_labels  = get_paths_and_labels(classes, "test")

train_ds = tf.data.Dataset.from_tensor_slices((train_paths, train_labels))
val_ds   = tf.data.Dataset.from_tensor_slices((val_paths, val_labels))
test_ds  = tf.data.Dataset.from_tensor_slices((test_paths, test_labels))


# ---- Safe loader ----
train_ds = (
    train_ds
    .map(safe_load_and_label, num_parallel_calls=AUTOTUNE)
    .filter(lambda img, label: tf.not_equal(label, -1))
)

val_ds = (
    val_ds
    .map(safe_load_and_label, num_parallel_calls=AUTOTUNE)
    .filter(lambda img, label: tf.not_equal(label, -1))
)

test_ds = (
    test_ds
    .map(safe_load_and_label, num_parallel_calls=AUTOTUNE)
    .filter(lambda img, label: tf.not_equal(label, -1))
)

# ---- Only TRAIN gets augmentation ----
train_ds = (
    train_ds
    .shuffle(1000)
    .map(tf_augment, num_parallel_calls=AUTOTUNE)  # Augment BEFORE batching
    .batch(BATCH_SIZE)
    .prefetch(AUTOTUNE)
)

# ---- VAL & TEST do NOT get augmentation ----
val_ds = val_ds.batch(BATCH_SIZE).prefetch(AUTOTUNE)
test_ds = test_ds.batch(BATCH_SIZE).prefetch(AUTOTUNE)

print("✅ Final dataset pipeline ready!")
```

```python
import random
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

# Extract clean validation paths from your val_paths TF tensor
val_paths_list = val_paths

def safe_load_pil(path):
    try:
        img = Image.open(path).convert("RGB")
        return img
    except:
        return None

def show_batch(n=4):
    picked = random.sample(val_paths_list, n)
    
    plt.figure(figsize=(12, 6))
    
    for i, path in enumerate(picked):
        img = safe_load_pil(path)
        if img is None:
            continue

        # Resize for display
        disp = img.resize((224, 224))

        plt.subplot(1, n, i+1)
        plt.imshow(disp)
        plt.title(Path(path).parent.parent.name)
        plt.axis("off")

    plt.tight_layout()
    plt.show()

show_batch(4)
```

```python
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np

sample_out = models_dir / "augment_samples"
sample_out.mkdir(exist_ok=True)

def save_aug_examples(n=4):
    for i, (img, label) in enumerate(train_ds.take(n)):
        # img is a batch; take the first
        arr = img[0].numpy()
        arr = (arr * 255).astype(np.uint8)
        name = f"aug_epoch_sample_{i}.jpg"
        Image.fromarray(arr).save(sample_out / name)
    print("✅ Augmentation samples saved!")

# call this after each epoch if you want (or only once)
```

```python
class SaveAugmentSamples(tf.keras.callbacks.Callback):
    def __init__(self, sample_path, out_dir, epochs):
        super().__init__()
        self.sample_path = sample_path
        self.out_dir = Path(out_dir)
        self.out_dir.mkdir(exist_ok=True)

        # Load original once
        self.original = Image.open(sample_path).convert("RGB")
        self.original = self.original.resize((IMG_SIZE, IMG_SIZE))

    def on_epoch_end(self, epoch, logs=None):
        epoch_id = epoch + 1

        # -- save original (A1) --
        orig_path = self.out_dir / f"epoch_{epoch_id:02d}_original.jpg"
        self.original.save(orig_path)

        # -- apply augmentation --
        img_np = np.array(self.original).astype(np.uint8)
        aug_np = augment_numpy(img_np)
        aug_img = Image.fromarray(aug_np)

        # -- save augmented (A1) --
        aug_path = self.out_dir / f"epoch_{epoch_id:02d}_augmented.jpg"
        aug_img.save(aug_path)

        # -- save side-by-side (A2) --
        combined = Image.new("RGB", (IMG_SIZE * 2, IMG_SIZE))
        combined.paste(self.original, (0, 0))
        combined.paste(aug_img, (IMG_SIZE, 0))

        combined_path = self.out_dir / f"epoch_{epoch_id:02d}_side_by_side.jpg"
        combined.save(combined_path)

        print(f"✅ Saved augmentation samples for epoch {epoch_id}")


# pick example validation image
sample_image_path = val_paths_list[0]

# create callback instance
save_aug_cb = SaveAugmentSamples(
    sample_path=sample_image_path,
    out_dir=augment_samples_dir,
    epochs=10
)
```

```python
from tensorflow.keras import layers, models

base = tf.keras.applications.MobileNetV3Small(
    input_shape=(IMG_SIZE, IMG_SIZE, 3),
    include_top=False,
    weights="imagenet",
    minimalistic=True
)

# Freeze base
base.trainable = False

inputs = layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3))
x = base(inputs, training=False)  # Remove preprocessing from model
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dropout(0.2)(x)
outputs = layers.Dense(num_classes, activation="softmax")(x)

model = models.Model(inputs, outputs)

model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-3),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

model.summary()
```

```python
EPOCHS = 10

lr_schedule = tf.keras.callbacks.ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=2,
    min_lr=1e-7
)

early_stop = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=3,
    restore_best_weights=True
)

history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS
    callbacks=[save_aug_cb, lr_schedule, early_stop]
)
```

```python
# Unfreeze and train with lower LR
base.trainable = True
model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-5),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

# Train 5 more epochs
history_fine = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=5
)
```

```python
test_loss, test_acc = model.evaluate(test_ds)
print(f"Test Accuracy: {test_acc:.4f}")
```

```python
model_save_path = models_dir / "plant_classifier_v1"
model.save(model_save_path)
print("Model saved to:", model_save_path)
```

```python
converter = tf.lite.TFLiteConverter.from_keras_model(model)
converter.optimizations = [tf.lite.Optimize.DEFAULT]
converter.target_spec.supported_types = [tf.int8]

def representative_dataset():
    for img, _ in val_ds.take(100):
        yield [img]

converter.representative_dataset = representative_dataset
converter.inference_input_type = tf.uint8
converter.inference_output_type = tf.uint8

tflite_model = converter.convert()

tflite_path = models_dir / "plant_classifier_quant.tflite"
tflite_path.write_bytes(tflite_model)
print(f"✅ Quantized TFLite model saved: {tflite_path}")
print(f"Model size: {len(tflite_model) / 1024:.2f} KB")
```
